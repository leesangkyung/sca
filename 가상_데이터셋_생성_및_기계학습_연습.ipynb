```python
import numpy as np
import matplotlib.pyplot as plt
import random as random

x = np.array([])
y = np.array([])

for i in range(0, 81, 1):
  t_i = int((i + 1) / 20) + 1
  x = np.append(x, random.randrange(t_i * 105, t_i * 111, 1))
  y = np.append(y, t_i)
print(x)
print(y)

plt.figure(figsize=(30, 20))
plt.scatter(x, y)
plt.xlabel("Sleep Time")
plt.ylabel("Sleep Cycle")
plt.title("test 1")
plt.show()

w = 0
b = 0

learning_rate = 0.000001
epochs = 100000

n = len(x)

for i in range(epochs):
  y_pred = w * x + b
  error = y - y_pred

  w_diff = -(2/n) * sum(x * (error))
  b_diff = -(2/n) * sum(error)

  w -= learning_rate * w_diff
  b -= learning_rate * b_diff

  if i % (epochs / 10) == 0:
    print(f"Epoch {i}: w = {w:.10f}, b = {b:.10f}")

plt.figure(figsize=(30, 20))
plt.scatter(x, y, color='blue')
plt.plot(x, w*x + b, color='red')
plt.xlabel("Sleep Time")
plt.ylabel("Sleep Cycle")
plt.title("final model")
plt.show()

# part II

```

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math as math

x = np.array([])
y = np.array([])
for i in range(1, 8, 1):
    for j in range(500 // (2 * i)):
        if j < 11:
            x1 = np.array([random.randrange(i * 60,100 * i)])
        elif j < 35:
            x1 = np.array([random.randrange(i * 70, 90 * i)])
        elif j < 70:
            x1 = np.array([random.randrange(i*75,85*i)])
        else:
            x1 = np.array([random.randrange(i*78,82*i)])
        y1 = np.array([i])
        x = np.append(x, x1)
        y = np.append(y, y1)


plt.figure(figsize=(9,6))
plt.scatter(x, y)
plt.xlabel('Sleeptime')
plt.ylabel('Sleep cycle')
plt.title('Sleep cycle / Time')
plt.show()
 # matplotlib

w = 0
b = 0

lr = 0.000001
epochs = 100000
n = len(x)

# 경사 하강법 시작
for i in range(epochs):
    y_pred = w * x + b  # 예측값 계산
    error = y - y_pred  # 오차 계산

    # 오차 함수의 w, b에 대한 편미분
    w_diff = -(2/n) * sum(x * (error))
    b_diff = -(2/n) * sum(error)

    # 파라미터 업데이트
    w -= lr * w_diff
    b -= lr * b_diff

    if i % (epochs / 10) == 0:  #실행 중 Epoch 별 출력해보기
        print(f"Epoch {i}: w = {w:.10f}, b = {b:.10f}")
# 최종 모델 시각화
plt.figure(figsize=(9, 6))
plt.scatter(x, y, color='blue')
plt.plot(x, w*x + b, color='red')   #linear regression 그리기
plt.xlabel('Sleeptime')
plt.ylabel('Sleep cycle')
plt.title('Sleep cycle / Time')
plt.show()
print(f"Final model parameters: w = {w:.10f}, b = {b:.10f}")
```
